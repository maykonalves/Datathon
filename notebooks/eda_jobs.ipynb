```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/raw/jobs.json')
```

```python
records = []
for key, value in df.items():
    record = {"id_vaga": key}
    for sub_key, sub_value in value.items():
        if isinstance(sub_value, dict):
            record.update(sub_value)
    records.append(record)

df = pd.DataFrame(records)

df.replace(['', ' ', '-', 'Nenhum -'], np.nan, inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"Column '{col}' has {df[col].nunique()} unique values.")
```

```python
for column in df.columns:
    if df[column].nunique() < 10:
        print(f"Column '{column}' unique values: {df[column].unique()}")
```

```python
df.rename(columns={'data_requicisao': 'data_requisicao'}, inplace=True)
```

```python
df['equipamentos_necessarios'] = df['equipamentos_necessarios'].apply(
    lambda x: x[:-2] if isinstance(x, str) and x.endswith(' -') else x
)
```

```python
df = df.drop(
    ['superior_imediato',
     'telefone',
     'pais',
     'solicitante_cliente',
     'requisitante',
     'nome',
     'regiao',
     'bairro',
     'nome_substituto',
     'horario_trabalho',
     'valor_compra_2',
     'analista_responsavel'
     ], axis=1)
```
